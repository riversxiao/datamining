## A very simple Recurrent Sequence

```python
def rnn(s1, w, length):
    s = s1
    result = [s1]
    for i in range(length):
        s = s * w - w * s**2 
        result.append(s)
    return result
    
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y = rnn(0.5,1,100)
x =list(range(101))
```
